```python
import numpy as np
import matplotlib.pyplot as plt

# Load spectral sensitivity variables into namespace
%run SpectralSensitivity.py

# ------------------------------------------------------------------
# Organize RGB data
# ------------------------------------------------------------------
channels = {
    "Blue":  blue,
    "Green": green,
    "Red":   red
}

colors = {
    "Blue":  "blue",
    "Green": "green",
    "Red":   "red"
}

# ------------------------------------------------------------------
# Plot raw (scaled) sensitivities
# ------------------------------------------------------------------
plt.figure(figsize=(9, 5))

for name, data in channels.items():
    plt.plot(wavelength_nm, data / 100, color=colors[name], label=name)

plt.xlabel("Wavelength (nm)")
plt.ylabel("Relative sensitivity")
plt.title("Spectral Sensitivity – RGB Channels (scaled)")
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------
# Normalization (GLOBAL maximum across all channels)
# ------------------------------------------------------------------
global_max = max(np.max(data) for data in channels.values())

channels_norm = {
    name: data / global_max
    for name, data in channels.items()
}

# ------------------------------------------------------------------
# Plot normalized sensitivities
# ------------------------------------------------------------------
plt.figure(figsize=(9, 5))

for name, data in channels_norm.items():
    plt.plot(wavelength_nm, data, color=colors[name], label=name)

plt.xlabel("Wavelength (nm)")
plt.ylabel("Normalized sensitivity")
plt.title("Normalized RGB Spectral Sensitivity")
plt.legend()
plt.grid(True)
plt.show()


```
