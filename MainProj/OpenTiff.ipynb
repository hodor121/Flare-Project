```python
import tifffile as tiff
import numpy as np 
import matplotlib.pyplot as plt
import os
folder_path = r"C:\Users\s234031\Desktop\1st_calibration"
tiff_files = [f for f in os.listdir(folder_path) if f.lower().endswith((".tiff", ".tif"))]

print("Tiff files found:")
for i, name in enumerate(tiff_files):
    print(f"{i}: {name}")
```

```python
tiff_file = os.path.join(folder_path, tiff_files[6])

image = tiff.imread(tiff_file)

print(image.shape, image.dtype)

plt.imshow(image, cmap="gray")
plt.title(tiff_files[6])
plt.axis("off")
plt.show()
```

```python
print("Min pixel value:", image.min())
print("Max pixel value:", image.max())
```

```python
plt.figure()
plt.hist(image.ravel(), bins=512)
plt.xlabel("Pixel value")
plt.ylabel("Count")
plt.title("Pixel intensity histogram")
plt.show()
```

```python
img_norm = (image - image.min()) / (image.max() - image.min())

plt.imshow(img_norm, cmap="gray")
plt.title(tiff_files[0])
plt.axis("off")
plt.show()
```
